# Problema del ejercicio b

Tiene como solución
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 $$
y tiene las siguientes condiciones de frontera:
$$u(0,y)=10y, ~~ u_x(1,y)=-1, \;\;\; 0 < y < 1$$
$$u(x,0)=u(x,1)=0\;\;\; 0 < x < 1 $$
Por lo tanto la serie de Fourier queda:
$$ u(x,y) = \sum_{n=1}^{\infty} \sin(n \pi y) [\frac{-20 (-1)^n}{n \pi} \cosh (n \pi x) + \frac{2[(-1)^n-1]+20(-1)^n(n \pi)\sinh (n\pi)}{(n \pi)^2 \cosh (n \pi)} \sinh(n \pi x)] $$


```python
from pylab import *
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

```python
def u_x_y(x,y):
    N = 8
    f = zeros ((len(x),len(y)))
    for i in range(len(x)):
        for n in range(1,N):
          a = (-20*(-1)^n)/(n*pi)
          b = (2*((-1)**n-1)+20*(-1)**n*(n*pi)*sinh(n*pi))/((n*pi)**2*cosh(n*pi))
          f[:,i] += sin(n*pi*y)*(a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))
    return f
```

Gráfica en 3D

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,13))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(Y, X, U, cmap="plasma"); #Otros colores:"viridis", "inferno", "plasma"
```
